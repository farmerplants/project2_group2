Transform Non-Stationary Data to Stationary

```python
# Convert dataframe to stationary data by taking the `pct_change` of the non-stationary data
march_df['march_data'] = march_df['march_data'].pct_change()
```

```python
# Drop the NaNs created by the pct_change function
march_df = march_df.dropna()
```

```python
# Plot the data as a line chart
march_df.march_data.plot()
```

```python
# Print the results of the AD Fuller test
result = adfuller(march_df.march_data)

# Access the contents of the results:
print('ADF Statistic: %f' % result[0])
print('p-value:        %f ' % result[1])
print('Lags used:      %d' % result[2])
print('Critical Values:')
for key, value in result[4].items():
    print((key, value))
if result[0] < result[4]["5%"]:
    print ("Reject Hypothesis - Time Series is Stationary")
else:
    print ("Failed to Reject Hypothesis - Time Series is Non-Stationary")
```

We see that the p-value is '0.00', and lower than 0.05, which means the series is stationary and is statistically significant. As a result, we can accept the Alternative Hypothesis.
