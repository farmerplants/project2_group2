Autocorrelation and Partial Autocorrelation Analysis

```python
# Plot the autocorrelation of the March and February five minute interval daily averages data
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
plot_acf(feb_march_df['five_minute_interval_avgs'], zero=False, lags=30)
```

The ACF appears to have a number of significant lags that slowly decline in signficance.

```python
# plot PACF to find AR lags `p`
plot_pacf(feb_march_df['five_minute_interval_avgs'], zero=False, lags=35)
```

The PACF appears to have 3 significant lags at 1, 2 and 3.  This means that most of our information is coming from these lags, but the other lags aren't exactly verbatim either and are contributing some weight as well.

As a result, we can tentatively estimate the order or our ARIMA model to be 3, 1, 1.


```python
# Print the results of the AD Fuller test
result = adfuller(feb_march_df.five_minute_interval_avgs)

# Access the contents of the results:
print('ADF Statistic: %f' % result[0])
print('p-value:        %f ' % result[1])
print('Lags used:      %d' % result[2])
print('Critical Values:')
for key, value in result[4].items():
    print((key, value))
if result[0] < result[4]["5%"]:
    print ("Reject Hypothesis - Time Series is Stationary")
else:
    print ("Failed to Reject Hypothesis - Time Series is Non-Stationary")
```
